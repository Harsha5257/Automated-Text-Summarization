# **Dataset**

```python
from datasets import load_dataset
multi_news = load_dataset("multi_news", split="test")
```

```python
multi_news.to_pandas()
```

**Dataset into Dataframe**

```python
documents = multi_news['document']
summaries = multi_news['summary']
```

```python
min_length = min(len(documents), len(summaries))
documents = documents[:min_length]
summaries = summaries[:min_length]
```

```python
import pandas as pd
```

```python
df = pd.DataFrame({"source_text": documents, "target_text": summaries})
```

```python
df.to_csv('/content/drive/MyDrive/Text Summarization/multinews.csv')
```

```python
df['source_text']="summarize: " + df['source_text']
```

**Train-Test Split**

```python
train_df = df.sample(frac=0.8, random_state=42)  # 80% for training
test_df = df.drop(train_df.index)  # Remaining for testing
```

```python
train_df.reset_index(drop=True, inplace=True)
test_df.reset_index(drop=True, inplace=True)
```

```python
train_df
```

# **Trainer**

```python
!pip install simplet5
```

```python
!pip install transformers==4.42.4
```

```python
from simplet5 import SimpleT5
```

```python
model = SimpleT5()
model.from_pretrained(model_type = "t5", model_name= "t5-base")
```

**Training**

```python
# train
model.train(train_df=train_df,
            eval_df=test_df,
            source_max_token_len = 512,
            target_max_token_len = 128,
            batch_size = 3,
            max_epochs = 3,
            use_gpu = True,
            outputdir = "/content/drive/MyDrive/Text Summarization/outputs",
            early_stopping_patience_epochs = 0,
            precision = 16
            )
```

```python
model.load_model("t5","/content/drive/MyDrive/Text Summarization/outputs/simplet5-epoch-2-train-loss-2.2298-val-loss-2.4041", use_gpu=True)
```

```python
text_summarize = """
Kedarnath: A helicopter, carrying 7, made an emergency landing due to a technical fault, approximately 100 metres away from the Kedarnath helipad. All the 6 passengers and the pilot are safe.
The video shows the helicopter tail spinning just before making an emergency landing, while people near the helipad scramble for cover.

"A helicopter of Kestrel Aviation co coming from Sirsi helipad to Sh Kedarnath Dham carrying 6 passengers along with a pilot had to make an emergency landing approx 100 metres before the helipad of Sh Kedarnath Dham at around 7.05 hrs due to some technical problem," district disaster management officer, Rudraprayag said.

This year, the Char Dham Yatra began on May 10, with the opening of three of the four shrines, including Gangotri, Yamunotri, and Kedarnath. The doors of Badrinath opened on May 12.

The Char Dham Yatra holds profound spiritual significance in Hinduism. This journey typically occurs from April-May to October-November.

It is believed that one should complete the Char Dham Yatra in a clockwise direction.

Hence, the pilgrimage starts from Yamunotri, proceeds towards Gangotri, onto Kedarnath, and finally ends at Badrinath.

"""

model.predict(text_summarize)
```

#**OCR**

```python
! pip install easyocr
```

```python
import easyocr
```

```python
reader = easyocr.Reader(['en'])
```

```python
def ocr_func(inp_image):
  ex_text = reader.readtext(inp_image,paragraph=True)

  for i in ex_text:
    in_text=i[1]

  text_sum = model.predict(in_text)
  for i in text_sum:
    if i =="['" or i =="']" or i == '["' or i == '"]':
      text_sum = text_sum.replace(i,"")

  return f'Summary:\n\n {text_sum}'
```

# **Gradio UI**

```python
! pip install gradio
import gradio as gr
```

```python
iface = gr.Interface(
    fn=ocr_func,
    inputs=gr.Image(label="Image Input"),
    outputs=['text'],
    title="Automated Text Summarization",
    description="Abstractive Text Summarization",

)

iface.launch()
```
